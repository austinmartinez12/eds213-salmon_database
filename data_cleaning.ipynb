```python
# Load libraries
import pandas as pd
import numpy as np
import duckdb
```

```python
# Read in data
crayfish = pd.read_csv("data/knb-lter-ntl.128.8/Biocom_pred_prey_crayfish.csv")
prey = pd.read_csv("data/knb-lter-ntl.128.8/Biocom_pred_prey_diet.csv")
fish = pd.read_csv("data/knb-lter-ntl.128.8/Biocom_pred_prey_fish.csv")
fishscales = pd.read_csv("data/knb-lter-ntl.128.8/Biocom_pred_prey_fishscales.csv")
```

```python
# Remove unececarry columns from prey data frame
# These are being removed becuase they are either too messy to work with or are not useful for answering my question
cols_to_drop = [
    "NOTES", "OTHER", "TERRESTRIAL_INVERTS", "EMERGENT_AQUATICS",
    "AMPHIPOD", "COLEOPTERANS", "HYDROCARINA", "CLADOCERA",
    "HEMIPTERA", "TRICHOPTERA", "DIPTERA", "NET_WT_OTHER_FISH",
    "OTHER_FISH", "MIMICSHINER", "BLUNTNOSE", "SMELT", "FISH",
    "CRAYFISH", "EPHEMEROPTERA", "ODONATA"
]

# Drop columns from prey
prey = prey.drop(columns=cols_to_drop)

# Fix the space in prey column name
prey = prey.rename(columns={'net_wt other': 'net_wt_other'})
```

```python
# Standardize all column names to lowercase, to make it easier to work with them
prey = prey.rename(columns=str.lower)
fish = fish.rename(columns=str.lower)
crayfish = crayfish.rename(columns=str.lower)
fishscales = fishscales.rename(columns=str.lower)
fish
```

```python
# Replace placeholder/missing value codes with NA
# This is very necessary for 'gutlabel' and 'scalelabel' because they are both primary keys
fish['gutlabel'] = fish['gutlabel'].replace('nolabel', pd.NA)
crayfish['gutlabel'] = crayfish['gutlabel'].replace('nolabel', pd.NA)
prey['scalelabel'] = prey['scalelabel'].replace({'nc': pd.NA, 'empty': pd.NA, 'na': pd.NA, 'rbx': pd.NA})
prey = prey.replace({'na': pd.NA, 'nc': pd.NA})

# Fix 'we007' typo in fish
# Database can't be created unless done
fish['gutlabel'] = fish['gutlabel'].replace('we007', 'we07')
```

```python
# Drop NAs in predator_fish["scalelabel"] and prey["gutlabel"]
# These will both be PKs and there can't be NAs in PKs
fishscales = fishscales.dropna(subset=["scalelabel"])
prey = prey.dropna(subset=["gutlabel"])

# Drop duplicates in predator_fish["scalelabel"] and prey["gutlabel"]
# These will both be PKs and there can't be duplicates in PKs
fishscales = fishscales.drop_duplicates(subset="scalelabel")
prey = prey.drop_duplicates(subset="gutlabel")
```

```python
# Export clean CSVs
prey.to_csv('data/clean_data/diet.csv', index=False)
fishscales.to_csv('data/clean_data/fishscales.csv', index=False)
fish.to_csv('data/clean_data/fish.csv', index=False)
crayfish.to_csv('data/clean_data/crayfish.csv', index=False)
```
