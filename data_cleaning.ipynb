```python
# Load libraries
import pandas as pd
import numpy as np
```

```python
# Read in data
crayfish = pd.read_csv("data/knb-lter-ntl.128.8/Biocom_pred_prey_crayfish.csv")
prey = pd.read_csv("data/knb-lter-ntl.128.8/Biocom_pred_prey_diet.csv")
fish = pd.read_csv("data/knb-lter-ntl.128.8/Biocom_pred_prey_fish.csv")
fishscales = pd.read_csv("data/knb-lter-ntl.128.8/Biocom_pred_prey_fishscales.csv")
```

```python
# Standardize all column names to lowercase
prey = prey.rename(columns=str.lower)
fish = fish.rename(columns=str.lower)
crayfish = crayfish.rename(columns=str.lower)
fishscales = fishscales.rename(columns=str.lower)
```

```python
# Strip whitespace from all string columns
cols = prey.select_dtypes(include="object").columns
for col in cols:
    prey[col] = prey[col].str.strip()

cols = fish.select_dtypes(include="object").columns
for col in cols:
    fish[col] = fish[col].str.strip()

cols = crayfish.select_dtypes(include="object").columns
for col in cols:
    crayfish[col] = crayfish[col].str.strip()

cols = fishscales.select_dtypes(include="object").columns
for col in cols:
    fishscales[col] = fishscales[col].str.strip()

# Fix the space in prey column name
prey = prey.rename(columns={'net_wt other': 'net_wt_other'})
```

```python
# Replace placeholder/missing value codes with NaN
fish['gutlabel'] = fish['gutlabel'].replace('nolabel', pd.NA)
crayfish['gutlabel'] = crayfish['gutlabel'].replace('nolabel', pd.NA)
prey['scalelabel'] = prey['scalelabel'].replace({'nc': pd.NA, 'empty': pd.NA, 'na': pd.NA, 'rbx': pd.NA})
prey = prey.replace({'na': pd.NA, 'nc': pd.NA})
```

```python
# Fix 'we007' typo in fish
fish['gutlabel'] = fish['gutlabel'].replace('we007', 'we07')

fishscales = fishscales.drop_duplicates(subset='scalelabel', keep='first')
```

```python
# Add synthetic IDs
fish['id'] = range(len(fish))
crayfish['id'] = range(len(crayfish))
#prey['id'] = range(len(prey))
```

```python
# applies to all count columns, not just odonata
count_cols = ['odonata', 'ephemeroptera', 'crayfish', 'fish', 'smelt', 'bluntnose', 
              'mimicshiner', 'other_fish', 'diptera', 'trichoptera', 'hemiptera', 
              'cladocera', 'hydrocarina', 'coleopterans', 'amphipod', 
              'emergent_aquatics', 'terrestrial_inverts', 'other']

for col in count_cols:
    prey[col] = pd.to_numeric(prey[col], errors='coerce')
```

```python
prey = prey.dropna(subset=['gutlabel'])
```

```python
prey = prey.drop_duplicates(subset='gutlabel', keep='first')
```

```python
# Export clean CSVs
prey.to_csv('data/clean_data/diet.csv', index=False)
fishscales.to_csv('data/clean_data/fishscales.csv', index=False)
fish.to_csv('data/clean_data/fish.csv', index=False)
crayfish.to_csv('data/clean_data/crayfish.csv', index=False)
```
